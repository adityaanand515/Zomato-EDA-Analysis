```python
# highlight features
df['highlight_list'] = df['highlights'].str.strip("[]").str.replace("'", "").str.split(', ')
exploded_df = df.explode('highlight_list')
# Groupby to get count and average rating
feature_stats = exploded_df.groupby('highlight_list').agg(
    feature_count=('highlight_list', 'count'),
    avg_rating=('aggregate_rating', 'mean')
).reset_index()
# 10 most frequent features
top_features = feature_stats.sort_values(by='feature_count', ascending=False).head(10)
# Plot feature distribution
plt.figure(figsize=(12, 6))
sns.barplot(data=top_features, x='feature_count', y='highlight_list', palette='Blues_r')
plt.title('Top 10 Most Common Restaurant Features')
plt.xlabel('Number of Restaurants')
plt.ylabel('Feature')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot feature vs rating
plt.figure(figsize=(12, 6))
sns.barplot(data=top_features.sort_values(by='avg_rating', ascending=False), 
            x='avg_rating', y='highlight_list', palette='coolwarm')
plt.title('Average Rating by Top Restaurant Features')
plt.xlabel('Average Rating')
plt.ylabel('Feature')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxenplot(df['average_cost_for_two'],color='teal', log_scale=True) 
plt.title('Distribution of Average Cost for Two (Log Scale)')
plt.xlabel('Average Cost (log scale)')
plt.ylabel('Number of Restaurants')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# The distribution shows that most restaurants offer meals for two at a low cost, primarily under ₹500. A few high-end restaurants exist, but they are rare outliers. The log scale reveals this sharp skew clearly.


```python
# Average cost by city
avg_cost_by_city = df.groupby('city')['average_cost_for_two'].mean().sort_values(ascending=False).head(10)

avg_cost_by_city.plot(kind='bar', figsize=(10,6), color='blue')
plt.title('Average Cost for Two by City')
plt.ylabel('Average Cost for Two')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
